
*  Custom Loop created to find the percentage of null values in each column

```python
# percentage of null values 

for i in df.columns:
    if df[i].isnull().sum():
        print(f"Percentage of null values in {i} : {(df[i].isnull().sum()/len(df))*100}")
    
```

```python
null_percent = (df.isnull().sum() / len(df)) * 100

null_percent = null_percent[null_percent > 0]

plt.pie(null_percent,labels=null_percent.index, autopct='%0.2f%%',explode=[0,1,0],shadow=True)
plt.title('Percentage of Non-Null Values in Each Column')
plt.show()
```

```python
null_counts = df.isnull().sum()

# bar plot
plt.bar(df.columns, null_counts)
plt.xlabel('Columns')
plt.ylabel('Number of Null Values')
plt.title('Count of Null Values in Each Column')
plt.xticks(rotation=45)
plt.show()
```

* Dropping Cabin Column as there are so many null values in Cabin
* we can adjust age column with mean values/scaling later

```python
df.drop('cabin',axis=1,inplace=True)
df
```

```python
# df[df.isnull().any(axis=1)]
df.isnull().sum()
```
